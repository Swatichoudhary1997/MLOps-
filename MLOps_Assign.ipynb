C. Demonstrate Model Stage Transitions
Use the MLflow Python API or the MLflow UI to transition the model through stages.

Transition to Staging: Set the initial registered version to the Staging environment.

```python
from mlflow.tracking import MlflowClient
client = MlflowClient()
model_version = model_details.version # Use the version registered in step A

client.transition_model_version_stage(
    name=model_name,
    version=model_version,
    stage="Staging"
)
print(f"Model {model_name} V{model_version} transitioned to Staging.")
```
